```python
import os
import torchvision.transforms as transforms
from PIL import Image
#https://pytorch.org/vision/stable/auto_examples/plot_transforms.html#sphx-glr-auto-examples-plot-transforms-py
```

```python
root_path=r'D:\youtube\2021\data_aug_pytorch\dogs-vs-cats\train'
img_path=os.path.join(root_path,'cat.15.jpg')
```

```python
img=Image.open(img_path)
img
```

```python
t1=transforms.RandomPerspective(distortion_scale=0.6, p=1.0)
t2=transforms.RandomRotation(degrees=(0, 180))
t3=transforms.RandomAffine(degrees=(30, 70), translate=(0.1, 0.3), scale=(0.5, 0.75))
t4=transforms.RandomCrop(size=(312, 312))

#t5=transforms.FiveCrop(224)

t_c=transforms.Compose([t4,t2])

img_1=t1(img)
img_2=t2(img)
img_3=t3(img)
img_4=t4(img)
img_c=t_c(img)
```

```python
print('transform_1: Random Perspective')
img_1
```

```python
print('transform_2: Random Rotation')
img_2
```

```python
print('transform_3: Random Affine')
img_3
```

```python
print('transform_4: Random crop')
img_4
```

```python
print('transform_compose: compose')
img_c
```
